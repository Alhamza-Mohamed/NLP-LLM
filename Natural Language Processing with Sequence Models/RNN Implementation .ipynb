```python
def scan(fn, elems, initializer = None):
    """
    fn: equivalent to Fw
    elems: set of all inputs x_t
    initializer: The hidden state h_t0
    """

    cur_value = initializer
    ys = []
    for x in elems:
        y, cur_value = fn(x, cur_value)
        ys.append(y)
    return ys, cur_value
```
