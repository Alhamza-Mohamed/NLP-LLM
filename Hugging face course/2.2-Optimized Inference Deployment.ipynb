```python
from huggingface_hub import InferenceClient

# Intialize client pointing to TGI endpoint

client = InferenceClient(base_url="http://localhost:8080", timeout=120)
#Use model= when Hugging Face hosts the model; use base_url= when you host the model.


```

## Text generation 

```python
out = client.text_generation(
    "Explain attention mechanisms in one paragraph.",
    max_new_tokens=120,
)

print(out)
```

## For chat format

```python
response = client.chat_completion( # send a chat-style generation request to the model
    messages = [        # provide the full conversation context as a list of message
        {"role": "system", "content":"You are a helpful assistant."}, # Define the assistant's behabior and rules
        {"role": "user", "content":"Tell me a story."} # Provide the user's prompt to the model
    ],
    max_tokens=100, # Limit the number of tokens generated in the response
    temperature = 0.7, # Control randomness (higher = more creative, lower = more deterministic)
    top_p = 0.95 #restrict token sampling to the most probable 95% mass
)
print (response.choices[0].message.content)
```

### using openai

```python
from openai import OpenAI

# Initialize client pointing to TGI endpoint
client = OpenAI(
    base_url="http://localhost:8080/v1",  # Make sure to include /v1
    api_key="not-needed",  # TGI doesn't require an API key by default
)

# Chat completion
response = client.chat.completions.create(
    model="HuggingFaceTB/SmolLM2-360M-Instruct",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Tell me a story"},
    ],
    max_tokens=100,
    temperature=0.7,
    top_p=0.95,
)
print (response.choices[0].message.content)
```

```python
from huggingface_hub import InferenceClient

client = InferenceClient(model="http://localhost:8081/v1" # URL to the llama.cpp server
                         , token="sk-no-key-required" # llama.cpp server requires this placeholder
                         )

# Advanced parameter example
response = client.chat_completion(
    messages = [
        {"role": "system", "content": "You are a creative storyteller."},
        {"role":"user","content":"Write a creative story"}
    ],
    temperature= 0.8,
    max_tokens=200,
    top_p = 0.95
)
print(response.choices[0].message.content)
```

```python
# For direct text generation 
# this code wont work as we use llama.cpp OpenAI-compatible chat API
# it needs Hugging Face text-generation API.
response = client.text_generation( 
    "Write a creative story about space exploration",
    max_new_tokens=200,
    temperature=0.8,
    top_p=0.95,
    repetition_penalty=1.1,
    details=True
) 
print (response.generated_text)
```

```python
# single user message (instead of normal text gereation)

response = client.chat_completion(
    messages=[
        {"role": "user", "content": "Write a creative story about space exploration"}
    ],
    temperature=0.8,
    max_tokens=200,
    top_p=0.95
)

print(response.choices[0].message.content)
```

```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:8081/v1",
    api_key="not-needed"
)

response = client.completions.create(
    model="local-model",
    prompt="Write a creative story about space exploration",
    max_tokens=200,
    temperature=0.8
)

print(response.choices[0].text)
```
