```python
import numpy as np 
```

## ReLU

```python
z = np.array([[-1.25459881], [ 4.50714306], [ 2.31993942], [ 0.98658484], [-3.4398136 ]])
z
```

```python
h = z.copy()
h[h<0] = 0 
h
```

```python
def relu (z):
    result = z.copy()
    result[result < 0] =  0
    return result
```

```python
print(relu(z))
```

## Softmax

```python
z = np.array([9, 8, 11, 10, 8.5])
```

```python
e_z = np.exp(z)
e_z
```

```python
sum_e_z = np.sum(e_z)
e_z[0]/sum_e_z
```

```python
def softmax(z):
    e_z = np.exp(z)
    sum_e_z = np.sum(e_z)
    return e_z/sum_e_z
```

```python
softmax([9, 8, 11, 10, 8.5])
```

```python
np.sum(softmax([9, 8, 11, 10, 8.5]))
```
